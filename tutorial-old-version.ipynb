Next, let us see the PCA loadings plot of `CA-125` and `Prolactin` levels in `Ovary` samples versus all the other cancer types

```python
cancer_PCA_reduced_list = []
cancer_explained_variance_list = []
cancer_components_df_list = []
cancer_n_components_list = [16, 19, 13, 14, 17, 21, 15, 15, 15]

for i in range(9):
    cancer_features, cancer_labels = feature_label_split(dfs[i])
    print(f"\nCategory: {categories[i]}")
    cancer_PCA_reduced, cancer_explained_variance, cancer_components_df = cancer_dataframe_PCA(features = cancer_features,
                                                                                               n_components = cancer_n_components_list[i],
                                                                                               scaler = StandardScaler())

    cancer_PCA_reduced_list.append(cancer_PCA_reduced)
    cancer_explained_variance_list.append(cancer_explained_variance)
    cancer_components_df_list.append(cancer_components_df)  
```

```python
cancer_explained_variance_list[0]
```

```python
def find_cancer_loadings(cancer_explained_variance, cancer_components_df):
    cancer_loadings_df = cancer_components_df * np.sqrt(cancer_explained_variance)
    return cancer_loadings_df

def plot_loadings_graphs_of_biomarker(cancer_explained_variance_ratio_list, cancer_components_df_list, biomarker_index):
    fig, axs = plt.subplots(3, 3, figsize=(15, 15), constrained_layout=True)
    axs = axs.flatten()
    for i in range(9):
        cancer_loadings_df = find_cancer_loadings(cancer_explained_variance_ratio_list[i], cancer_components_df_list[i])
        biomarker_loading_plot_df = pd.DataFrame()
        biomarker_loading_plot_df['pc'] = [i+1 for i in range(len(cancer_loadings_df.columns))]
        biomarker_loading_plot_df['loading'] = cancer_loadings_df.iloc[biomarker_index].values
        sns.lineplot(data = biomarker_loading_plot_df, x='pc', y='loading', ax=axs[i])
        axs[i].set_xticks(biomarker_loading_plot_df['pc'])
        axs[i].set_xlabel(f'Principal Components ({len(cancer_loadings_df.columns)})')
        axs[i].set_ylabel('Loading')
        axs[i].set_title(f'{categories[i]}')
    fig.suptitle(f'{biomarkers[biomarker_index]} Loadings')
    
plot_loadings_graphs_of_biomarker(cancer_explained_variance_list, cancer_components_df_list, 3)
```
